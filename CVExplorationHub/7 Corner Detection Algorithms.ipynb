"Corner Detection Algorithms Implementation for Image Analysis"

This notebook explores and implements popular corner detection algorithms, including Harris Corner, Shi-Tomasi, SIFT (Scale-Invariant Feature Transform), and SURF (Speeded-Up Robust Features). Through these implementations, we aim to detect and highlight key points or corners within images, providing insights into the distinctive features present. The analysis showcases the diverse applications of corner detection in computer vision and image processing, offering a comprehensive guide for practitioners and researchers in the field.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("wanted.jpg")
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray,cmap = "gray")
```

## Harris Corner Detection

Harris Corner Detection: Identifies corners based on local intensity variations, highlighting points where small movements in any direction result in significant changes in image structure.

```python
har = cv2.cornerHarris(image_gray,2,3,0.04)
har = cv2.dilate(har,None)
image[har > 0.01 * har.max()]=[0,0,255]
```

```python
plt.imshow(har)
```

```python
plt.imshow(image)
```

## Shi-Tomasi Corner Detection

Shi-Tomasi Corner Detection: An improvement over Harris Corner, Shi-Tomasi selects corners based on the minimum eigenvalue, offering better performance and stability in identifying significant features.

```python
import numpy as np
image = cv2.imread("wanted.jpg")
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
corners = cv2.goodFeaturesToTrack(image_gray,25,0.01,10)
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(image,(x,y),3,255,-1)
```

```python
plt.imshow(image)
```

## SIFT

SIFT (Scale-Invariant Feature Transform): A robust feature extraction method that detects and describes key points in images while maintaining scale invariance, enabling reliable matching across different scales and orientations.

```python
import numpy as np
image = cv2.imread("wanted.jpg")
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
sift = cv2.SIFT_create()
kp = sift.detect(image_gray,None)
```

```python
image = cv2.drawKeypoints(image_gray,kp,image)
```

```python
plt.imshow(image)
```
