"Image Blending and Alpha Masking in OpenCV: A Python Implementation"

This notebook demonstrates the process of blending two images using alpha blending, creating a visually appealing composition. the notebook explores alpha masking by reading an input image and applying a mask to selectively reveal or hide portions of the image, offering insights into advanced image manipulation techniques.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("/content/IMG_20220808_200225.jpg")
# img1 = cv2.cvtColor(img1,cv2.COLOR_BAYER_BG2RGB)
img2 = cv2.imread("/content/painting.jpg")
```

```python
img1 = cv2.resize(img1, (255,255), interpolation = cv2.INTER_AREA)
img2 = cv2.resize(img2, (255,255), interpolation = cv2.INTER_AREA)
```

```python
img1.shape
img2.shape
```

Alpha Blending

Alpha Blending:
Alpha blending combines two images by assigning weight values (alpha and beta) to each, producing a seamless blend that gives the impression of one image overlapping the other. The OpenCV function facilitates this process, allowing control over the intensity of each image's contribution.

```python
alpha = 0.8
beta = (1.0-alpha)
dst = cv2.addWeighted(img1,alpha,img2,beta,0.0)
```

```python
plt.imshow(dst)
plt.show()
```

Masking

Alpha Masking:
Alpha masking involves selectively revealing or hiding portions of an image using a mask. By applying an alpha channel to the mask, transparency levels control the visibility of specific regions in the input image, enabling precise control over the visual composition.

```python
im = cv2.imread("/content/painting.jpg", cv2.IMREAD_UNCHANGED)
```

```python
_, mask = cv2.threshold(im[:, :], 0, 255, cv2.THRESH_BINARY)
```

```python

plt.imshow(mask)
plt.show()
```
