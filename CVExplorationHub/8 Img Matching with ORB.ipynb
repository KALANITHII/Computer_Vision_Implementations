"Image Matching with ORB Descriptors Using Brute-Force Method in OpenCV"

This notebook demonstrates image matching between an input or query image and a training image or scene using the Brute-Force method with ORB (Oriented FAST and Rotated BRIEF) descriptors in OpenCV. The program reads the input image, extracts ORB descriptors for keypoints in both images, performs matching, and displays the top matched keypoints, providing a visual representation of their correspondence. This tutorial serves as a practical guide for implementing image matching techniques in computer vision applications using OpenCV.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_1 = cv2.imread("/content/Spiderverse.jpg")
```

```python
image_1_gray = cv2.cvtColor(image_1,cv2.COLOR_BGR2GRAY)
```

```python
xxx=cv2.rotate(image_1,cv2.ROTATE_90_CLOCKWISE)
```

```python
# image_2 = cv2.imread("./Driver Gosling - Rotated.jpg")
```

```python

image_2_gray = cv2.cvtColor(xxx,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_1_gray,cmap = "gray")
```

```python
plt.imshow(image_2_gray,cmap = "gray")
```

```python
# ORB Detector

detect = cv2.ORB_create()
key_point1,descrip1 = detect.detectAndCompute(image_1_gray,None)
key_point2,descrip2 = detect.detectAndCompute(image_2_gray,None)
```

```python
# BF
brute_force = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
no_of_matches = brute_force.match(descrip1,descrip2)

no_of_matches = sorted(no_of_matches,key = lambda x:x.distance)
```

```python
tot_feature_matches = len(no_of_matches)
tot_feature_matches
```

```python
output_image = cv2.drawMatches(image_1_gray,key_point1,image_2_gray,key_point2,no_of_matches[:30],None,flags = 2)
```

```python
plt.imshow(output_image)
```
